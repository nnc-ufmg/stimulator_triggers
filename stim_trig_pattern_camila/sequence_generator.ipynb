```python
import random

def generate_numbers_with_1s(X):
    X = X + 1
    results = []
    for i in range(2**(X-2), 2**(X-1)):
        binary = bin(i)[2:]
        if binary[0] == '1' and binary[-1] == '1':
            results.append(binary)
    return results

number_of_bins = 10
number_of_stimuli = 6

numbers = generate_numbers_with_1s(number_of_bins)
print("Números binários de", number_of_bins, "unidades com MSB e LSB iguais a 1:")

numbers_stim = []
for number in numbers:
    ones = sum([int(x) for x in number])
    if ones == number_of_stimuli:
        numbers_stim.append(number)

print(len(numbers_stim), "números com", number_of_stimuli, "unidades de 1:")
if len(numbers_stim) < 210:
    missing = 210 - len(numbers_stim)
    repetition = missing // len(numbers_stim)
    numbers_stim = numbers_stim * int(repetition)
    
    missing = 210 - len(numbers_stim)
    numbers_stim += numbers_stim[:missing]

random.seed(10)
random.shuffle(numbers_stim)

numers_copy = ",".join(numbers_stim)
print(numers_copy)

in_decimal = [int(x, 2) for x in numbers_stim]
print("\nNúmeros em decimal:")

decimal_copy = ",".join(map(str, in_decimal))
print(decimal_copy)
```
